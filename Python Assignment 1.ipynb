```python
text = "Hello Ayisha, welcome to Python programming"

print("First character:", text[0])
print("Last character:", text[-1])
print("First 5 characters:", text[:5])
print("Last 11 characters:", text[-11:])
print("Reversed:", text[::-1])
print("Extracted word:", text[text.find("Python"):text.find("Python")+6])
```

```python
strM = "Python beginner tutorial"

print(strM.upper())
print(strM.lower())
print(strM.capitalize())
print("Count of 't':", strM.count('t'))
print(strM.replace("Python", "Machine Learning"))
```

```python
t1 = (10, 20, 30)
t2 = (40, 50, 60)

t_combine = t1 + t2
print("Combined:", t_combine)

print("Repeated:", t_combine * 3)

print("3rd element:", t_combine[2])

print("First three:", t_combine[:3])

print("Last three:", t_combine[-3:])
```
